```python
import pandas as pd
```

```python
articles = pd.read_csv("/content/drive/MyDrive/Group 1/articles.CDKN2A.csv")
articles
```

```python
#select the last 5 years from 'articles'
#set the 'cutoff' date so the first of 5 years
cutoff_year = pd.to_datetime('2019-01-01')
articles_5 = articles[articles['Year'] >= cutoff_year]
articles_5
```

```python
#FirstAuthorForename articles_5 FirstAuthorLastname

all_text = ' '.join(articles_5['FirstAuthorLastname'].dropna().astype(str))

words = all_text.split()
word_series = pd.Series(words)
word_counts = word_series.value_counts()
most_common_word = word_counts.idxmax()
print(f"The most common word is: {most_common_word}")
top_5_words = word_counts.head(5)
print(top_5_words)

```

```python
authors = pd.read_csv("/content/drive/MyDrive/Group 1/authors.CDKN2A.csv")
authors
```
